```
# Data Visualization
# It is easier to visualize data instead of reading it 
# Data can be Vizualized in many forms like graphs, plots etc.
# we are here using matplotlib 
```

```
#importing pyplot library under matplotlib module
import matplotlib.pyplot as plt
```

```
x=[12,5]
y=[5,12]
plt.xlabel("time")
plt.ylabel("Distance")
plt.plot(x,y)
plt.grid(color='green')     # to show the grid lines of green color 
plt.show()                   # to show the 
```

```
x=[12,25]
y=[25,12]
x2=[9,18]
y2=[28,9]
plt.xlabel("time")
plt.ylabel("Distance")
plt.plot(x,y,label="speed of car")     # we write the labels because we want the legends to be displayed
plt.plot(x2,y2,label="speed of airplane")
plt.grid(color='green')     # to show the grid lines of green color 
plt.legend()                 # to print the legends
plt.show()                   # to show the graph
```
